```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = [
    Point(xy)
    for xy in zip(geolocation_dataset["Longitude"], geolocation_dataset["Latitude"])
]
gdf = GeoDataFrame(geolocation_dataset, geometry=geometry)
germany = gpd.read_file("data/germany_geolocation.json")
gdf.plot(ax=germany.plot(figsize=(20, 20)), marker=".", color="red", markersize=100)
```
